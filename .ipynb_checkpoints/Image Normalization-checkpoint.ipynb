```python
#import libaries
import numpy as np
import pandas as pd
import re
import cv2
from PIL import Image
import matplotlib.pyplot as plt
plt.style.use('seaborn-darkgrid')
%matplotlib inline

#Misc for ignoring warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
#Load random cropped image
Image.open("./pre_cropped/crop_part1/1_0_0_20161219163425847.jpg.chip.jpg")
```

```python
#define the cropped images path with variable faces
faces = "./pre_cropped/crop_part1"
```

### Image Normalization

```python
face = [] #empty list for appending values
image_size = 64 #resize to 64*64
for image in (os.listdir(faces)): #read the files
    path = os.path.join(faces, image) #assign the image
    img = cv2.imread(path, cv2.IMREAD_GRAYSCALE) #convert to gray scale
    img = cv2.resize(img, (image_size, image_size)).flatten()  #resize and flatten the image 
    face.append(img) #append img
```

```python
X = np.array(face) #converting to array
X.shape,X.flatten().shape #checking shape
```

### Extracting lables from the image file name

File name as different coding like **0 > male** , **1 > female**, but capstone conditions are to use 0 for female and 1 for male. So Used if else condition for it

```python
labels = [] #empty list
for name in (os.listdir(faces)): #read the files
    re_name = re.sub(r"_[0|1|2|3|4]_\d{17}.jpg.chip.jpg","",name) #regx for getting the label
    re_label = re_name.split(sep="_")[-1] #seprating the left out part
    gen = int(re_label) #string to int
    if gen==1: #capstone conditions
        labels.append(0)
    else:
        labels.append(1)
```

```python
y = np.array(labels) #converting to array
y.shape #checking shape
```

```python
gen,count = np.unique(y,return_counts=True)
dict(zip(gen, count)) #checking females and males
```
